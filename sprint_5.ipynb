## Step 2: Data Preparation

### 2.1 Standardizing Column Names

````python
# Convert column names to lowercase
ice_df.columns = ice_df.columns.str.lower()
````

````python
# Verify the changes
print(ice_df.columns)
````

### 2.2 Data Type Conversion

````python
# Check current data types
ice_df.dtypes
````

````python
# Change year_of_release from decimal to Int64 - cleaner to read 
ice_df['year_of_release'] = ice_df['year_of_release'].astype('Int64')

# Change user_score to numeric value (object --> float) for consistency and replace tbd with np.nan for missing values
ice_df['user_score'] = pd.to_numeric(ice_df['user_score'].replace('tbd', np.nan), errors='coerce')

# Convert columns with repeated text to category since data isn't numerical & it will be easier when comparing - platform, genre, rating
ice_df['platform'] = ice_df['platform'].astype('category')
ice_df['genre'] = ice_df['genre'].astype('category')
ice_df['rating'] = ice_df['rating'].astype('category')

# Check dtypes
ice_df.dtypes
````

## 3.3 Determining Relevant Time Period

Based on your analysis above, determine the appropriate time period for predicting 2017 sales:

<div style="background-color: #ADD8E64D; padding: 15px; margin: 10px;">

**Why 2013‚Äì2016 Is the Most Relevant Period**

1. Reflects the Current Console Generation

  - Major platforms like PlayStation 4 (PS4) and Xbox One (XOne) launched in late 2013.

  - These years capture the entire growth phase of the newest consoles available at the time.

  - Shows a range, so that the data isn't skewed; older consoles like PS3, Xbox 360 are in visible decline, while newer ones are still rising.


2. Most Relevant for 2017 Forecasting

  - To forecast 2017 trends, we want data from recent, stable, and comparable years.

  - Sales patterns from 2013‚Äì2016 show current consumer behavior, platform performance, and genre interest.

  - Including older years would bring in outdated platforms no longer relevant to 2017.


3. Shift in Market Dynamics

  - The heatmap shows a hand-off between console generations; where PS3/X360 fade out, and PS4/XOne rise.

  - 2013‚Äì2016 most cleanly captures that shift ‚Äî making it ideal for identifying current trends.


4. Data Consistency and Reliability

  - Older years may have less consistent data or missing review scores.

  - Recent years tend to have more complete records, making the analysis more accurate.



In summary: Focusing on 2013‚Äì2016 ensures our analysis is based on platforms that are active, growing, and most likely to matter in 2017. It avoids outdated trends and gives a reliable foundation for sales forecasting and marketing strategy.

````python
# Filter data for years 2013‚Äì2016
filtered_df = ice_df[ice_df['year_of_release'].between(2013, 2016)]

# Group total sales by platform and year
platform_year_sales = (
    filtered_df.groupby(['platform', 'year_of_release'])['total_sales']
    .sum()
    .reset_index()
)

# Get top 10 platforms based on total sales
top_platforms = (
    filtered_df.groupby('platform')['total_sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)  # Limit to top 10
    .reset_index()
)

# Reorder pivot to match top platforms
platform_order = top_platforms['platform'].tolist()

# Create pivot table
sales_pivot = platform_year_sales[
    platform_year_sales['platform'].isin(platform_order)
].pivot(index='platform', columns='year_of_release', values='total_sales').fillna(0) 

# Filter platform_order to only include platforms that actually appear in the pivot
available_platforms = sales_pivot.index.tolist()
platform_order = [p for p in platform_order if p in available_platforms]

# Reindex rows to match order
sales_pivot = sales_pivot.reindex(platform_order)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    sales_pivot,
    cmap='Blues',
    annot=True,
    fmt='.1f',
    annot_kws={'size': 10},
    linewidths=0.5,
    cbar_kws={'label': 'Sales (Millions USD)'}
)

plt.title('Top 10 Platform Sales by Year (2013‚Äì2016)', fontsize=16, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Platform', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10, rotation=0)
plt.tight_layout()
plt.show()

# Summary Stats
print(f"Total platforms analyzed: {len(platform_order)}")
print(f"Years covered: {sorted(sales_pivot.columns.tolist())}")
print("Top 3 platforms by total sales:")
for i, row in top_platforms.head(3).iterrows():
    print(f"  {row['platform']}: {row['total_sales']:.1f}M USD")
````

## 4.1 Regional Platform Analysis

Let's begin by examining platform performance across different regions:

````python

# Function to analyze platform performance by region NA, EU, JP
def calculate_top5_platforms_by_region(df, sales_column, region_name):
    # Group by platform and sum the sales for the selected region
    region_sales = (
        df.groupby('platform')[sales_column]
        .sum()
        .reset_index()
        .sort_values(by=sales_column, ascending=False)
    )

    # Calculate total regional sales to get market share
    total_regional_sales = region_sales[sales_column].sum()
    region_sales['market_share'] = ((region_sales[sales_column] / total_regional_sales) * 100).round(2)

    # Get top 5 platforms
    top5 = region_sales.head(5).reset_index(drop=True)

    # Format output
    top5_display = top5.copy() # copy df to avoid modifying original
    top5_display.columns = ['Platform', f'{sales_column.title()}', 'Market Share (%)']
    print(f"\nTop 5 Platforms in {region_name}")
    print(top5_display.to_string(index=False, float_format='%.2f'))

    return top5_display 
````

## 4.3 ESRB Rating Impact Analysis

Finally, let's examine how ESRB ratings affect sales in each region:

````python
# Function to analyze ESRB rating impact

def calculate_esrb_by_region(df, esrb_column, region_name):
    # Group by rating and sum ESRB for the selected region & sort highest to lowest
    region_esrb = (
        df.groupby('rating')[esrb_column]
        .sum()
        .reset_index()
        .sort_values(by=esrb_column, ascending=False)
    )

    # Get total sales ranked by regional ESRB & calculate market share
    total_sales_by_esrb = region_esrb[esrb_column].sum()
    region_esrb['market_share'] = (region_esrb[esrb_column] / total_sales_by_esrb * 100).round(2)

    # Format column names for consistency
    region_esrb.columns = ['ESRB Rating', 'Sales', 'Market Share (%)']

    # Display results & round decimal
    print(f"\nESRB Rating Breakdown in {region_name}")
    print(region_esrb.to_string(index=False, float_format="%.2f"))

    return region_esrb
````

````python
## Run ESRB rating analysis for each region (NA, EU, JP)

na_esrb = calculate_esrb_by_region(filtered_df, 'na_sales', 'North America')
eu_esrb = calculate_esrb_by_region(filtered_df, 'eu_sales', 'Europe')
jp_esrb = calculate_esrb_by_region(filtered_df, 'jp_sales', 'Japan')
````

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

‚ÄîHow you formulated the null and alternative hypotheses

‚ÄîWhat criteria you used to test the hypotheses~~,~~ and why


<div style="border: 4px solid lightblue; padding: 15px; margin: 10px;">

**Question 1:** Are average user ratings of Xbox One and PC platforms the same?
  
  * Implementing a t-test because we want to compare the means of two independent groups. This is a two-tailed hypothesis, with the aim to test for any difference. 
  * Alpha will be 0.05 as this is a common standard for establishing a balance between avoiding false negatives and false positives - meaning there is a 5% chance of committing a Type 1 error. 

    1. Null Hypothesis: the average user ratings for Xbox One and PC are the same.
    2. Alternative Hypothesis: the average user ratings for Xbox One and PC are not the same.  

</div>

<div style="border: 5px solid green; padding: 10px; margin: 5px">
<b> Reviewer's comment </b>

Good. The null hypothesis always includes the equality sign. This is because the test does not understand exactly how we set the problem: we can say, for instance, let's make sure that they are not equal. Or that they are equal. And regardless of the purpose of our study, we always put equality at null hypothesis. Then the test result is interpreted correctly.

</div>   

````python
from scipy.stats import ttest_ind

# Filter user scores for Xbox One and PC
xbox_user_scores = filtered_df[(filtered_df['platform'] == 'XOne')]['user_score'].dropna()
pc_user_scores = filtered_df[(filtered_df['platform'] == 'PC')]['user_score'].dropna()

# Conduct independent t-test
t_stat, p_value = ttest_ind(xbox_user_scores, pc_user_scores, equal_var=False)  # Welch's t-test

# Results
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in average user scores between Xbox One and PC.")
else:
    print("Fail to reject the null hypothesis: No significant difference in average user scores between Xbox One and PC.")
````

**Xbox 1 vs. PC User Ratings:**

- T-statistic of 1.45 suggests the difference in average user ratings between platforms is small and not far from what we'd expect by chance.
- The p-value of 0.15 means there's a 15% probability of seeing a difference like this if there‚Äôs actually no real difference ‚Äî which is not low enough to be considered statistically significant.

**Findings:** Users rate games similarly on both platforms. We cannot conclude a meaningful difference in user sentiment, so it‚Äôs best to focus on sales performance and market share by region when making platform-specific advertising decisions.

**Question 2:** Are average user ratings for Action genres different from average user ratings for Sports genres?
  
  * Implementing a t-test because we want to compare the means of two independent groups. This is a two-tailed hypothesis, with the aim to test for any difference. 
  * Alpha will be 0.05 as this is a common standard for establishing a balance between avoiding false negatives and false positives - meaning there is a 5% chance of committing a Type 1 error. 

    1. Null Hypothesis: The average user ratings for Action and Sports games are the same.
    2. Alternative Hypothesis: The average user ratings for Action and Sports games are differen

````python
# Filter user scores for Action and Sports
action_ratings = filtered_df[filtered_df['genre'] == 'Action']['user_score'].dropna()
sports_ratings = filtered_df[filtered_df['genre'] == 'Sports']['user_score'].dropna()

# Conduct independant t-test
t_stat, p_value = ttest_ind(action_ratings, sports_ratings)

# Results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
````

````python
# Calculate mean user scores for Action and Sports genres
action_mean = filtered_df[filtered_df['genre'] == 'Action']['user_score'].dropna().mean()
sports_mean = filtered_df[filtered_df['genre'] == 'Sports']['user_score'].dropna().mean()

print(f"Average user score for Action games: {round(action_mean, 2)}")
print(f"Average user score for Sports games: {round(sports_mean, 2)}")
````

**Findings:**
